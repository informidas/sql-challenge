```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sqlalchemy import create_engine
from config import auth_user
```

```python
# set db connection string
db_string = f"postgresql://{auth_user}@localhost:5432/PewlettHackard"
```

```python
# instantiate connection
engine = create_engine(db_string)
```

```python
# execute query against employees table 
emp_data = engine.execute("SELECT * FROM employees")

# create an employee list
employees = []

# append records returned from db to the emp_data object
for record in emp_data:
    employees.append(record)
```

```python
# generate an employees dataframe
employees_df_raw = pd.DataFrame(employees)
employees_df_raw
```

```python
# rename column headings in the employees dataframe
employees_df_renamed = employees_df_raw.rename(columns={0:"emp_no", 1:"birth_date", 2:"first_name", 3:"last_name",4:"gender", 5:"hire_date"})
employees_df_renamed
```

```python
# execute query against salaries table 
sal_data = engine.execute("SELECT emp_no, salary, from_date, to_date FROM salaries")

# create a salaries list
salaries = []

# use for loop to append records returned from db to the sal_data object
for salary in sal_data:
    salaries.append(salary)
```

```python
# generate a salaries dataframe
salaries_df_raw = pd.DataFrame(salaries)
salaries_df_raw
```

```python
# rename column headings in the salaries dataframe
salaries_df_renamed = salaries_df_raw.rename(columns={0:"emp_no", 1:"salary", 2:"from_date", 3:"to_date"})
salaries_df_renamed
```

```python
# execute query against employees table 
title_data = engine.execute("SELECT * FROM titles")

# create an employee list
titles = []

# append records returned from db to the emp_data object
for record in title_data:
    titles.append(record)
```

```python
# generate a titles dataframe
titles_df_raw = pd.DataFrame(titles)
titles_df_raw
```

```python
# rename column headings in the titles dataframe
titles_df_renamed = titles_df_raw.rename(columns={0:"emp_no", 1:"title", 2:"from_date", 3:"to_date"})
titles_df_renamed
```

```python
# combine employees and salaries dataframe 
employees_salary_df_combined = pd.merge(employees_df_renamed, salaries_df_renamed, on="emp_no", how="inner")
employees_salary_df_combined.head()
```

```python
employees_salary_df = employees_salary_df_combined.sort_values(by=['salary'])
employees_salary_df
```

```python
# Check to see if the salaries are normally distributed and for outliers

salary = employees_salary_df['salary']

fig1, ax1 = plt.subplots()
ax1.set_title('Salaries at Pewlett Packard')
ax1.set_ylabel('Salary ($000s)')
ax1.boxplot(salary)
plt.show()
```

```python
# show quartile calculations
quartiles = salary.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile for salary is: {lowerq}")
print(f"The upper quartile for salary is: {upperq}")
print(f"The interquartile range for salary is: {iqr}")
print(f"The the median for salary is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# create bins for data
bins= [39999, 49999,59999, 69999, 79999, 89999, 99999, 109999, 119999, 129999]
group_labels = ["40 to 49k","50 to 59k","60 to 69k","70 to 79k","80 to 89k","90 to 99k","100 to 109k","110 to 119k","120 to 129k"]
```

```python
# slice data against the bins
employees_salary_df['salary_range'] = pd.cut(employees_salary_df['salary'], bins, labels=group_labels)
employees_salary_df.head()
```

```python
# Create a GroupBy object based upon "salary_range"
salary_group = employees_salary_df.groupby("salary_range")

# Find how many rows fall into each bin
salary_ranges_df_raw = pd.DataFrame(salary_group["emp_no"].count()).reset_index()
salary_ranges_df = salary_ranges_df_raw.rename(columns={"emp_no":"count_of_employees"})
salary_ranges_df
```

```python
# plot Histogram for Salary Distribution
plt.figure(figsize=[13,7])
plt.hist(employees_salary_df['salary'], bins=10, color='#0504aa',alpha=0.8, rwidth=0.85)
plt.title('Salary Distribution Histogrm for Pewlett Hackard Employees',fontsize=15)
plt.xlabel('Salary',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
# get a filtered version of employees_salary_df to combine with titles dataframe
emp_salary_filtered = pd.DataFrame(employees_salary_df[['emp_no', 'salary']]).reset_index()
del emp_salary_filtered['index']
emp_salary_reduced_df = pd.DataFrame(emp_salary_filtered)
emp_salary_reduced_df
```

```python
# combine employees and titles dataframes 
employees_title_df = pd.merge(emp_salary_reduced_df, titles_df_renamed, on="emp_no", how="inner")
employees_title_df.head()
```

```python
# Create a GroupBy object based upon "Title"
title_group = employees_title_df.groupby("title")

# Find average salary for each title
salary_title_df_raw = pd.DataFrame(title_group["salary"].mean()).reset_index()
salary_title_df_renamed = salary_title_df_raw.rename(columns={"salary":"average_salary"})
salary_title_df_renamed["average_salary"] = salary_title_df_renamed["average_salary"].round()
salary_title_df = salary_title_df_renamed.sort_values(by=['average_salary']).reset_index()
del salary_title_df['index']
salary_title_df
```

```python
# create list for x plots
titles = salary_title_df["title"].to_list()
```

```python
# create list for y plots
avg_salaries = salary_title_df["average_salary"].to_list()
```

```python
# Create a bar chart based upon the above data
plt.figure(figsize=(20,10))
# Give the chart a title, x label, and y label
plt.title("Average Salary By Title for Pewlett Packard Employees" ,fontsize=15)
plt.xlabel("Title", fontsize=15)
plt.ylabel("Count of Employees", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', alpha=0.5)
plt.bar(titles,avg_salaries , color="b", align="center")
```
